```python
import os,ogr
path = os.getcwd()+"/shp/"
```

```python
import gdal
gdal.SetConfigOption("GDAL_FILENAME_IS_UTF8", "YES")
gdal.SetConfigOption("SHAPE_ENCODING", "CP936")
```

```python
pnt = path + "北京_point.shp"
driver = ogr.GetDriverByName('ESRI Shapefile')
dataSource = driver.Open(pnt, 0)
```

```python
layer = dataSource.GetLayerByIndex(0)
```

```python
lyd = layer.GetLayerDefn()
fields = [lyd.GetFieldDefn(i).GetName() for i in range(lyd.GetFieldCount())]
```

```python
fields
```

```python
wkt = "POLYGON((116.300784 40.005689,\
                116.331398 40.005523,\
                116.333698 39.986619,\
                116.304808 39.987559,\
                116.300784 40.005689))"
```

```python
sf = ogr.CreateGeometryFromWkt(wkt)
```

```python
print(sf)
```

```python
layer.SetSpatialFilter(sf)
feDict = []
for feature in layer:
    name = feature.GetField("NAME")
    print("{0}".format(name))
    f = {}
    for field in fields:
        f[field] = feature.GetField(field)
    geo = feature.GetGeometryRef()
    f["Shape"] = geo.ExportToWkt()
    feDict.append(f)
```

```python
import pandas as pd
```

```python
sfpd = pd.DataFrame(feDict)
sfpd
```

```python
bj = sfpd[sfpd["NAME"].str.contains(r'北京大学*')]
bj
```

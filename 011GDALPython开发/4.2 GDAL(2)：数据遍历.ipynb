```python
import os,ogr
path = os.getcwd()+"/shp/"
```

```python
xzqh = path + "行政区划.shp"
```

```python
driver = ogr.GetDriverByName('ESRI Shapefile')
dataSource = driver.Open(xzqh, 0)
```

```python
layer = dataSource.GetLayerByIndex(0)
```

```python
lyd = layer.GetLayerDefn()
fields = [lyd.GetFieldDefn(i).GetName() for i in range(lyd.GetFieldCount())]
```

```python
fields
```

```python
for i in range(layer.GetFeatureCount()):
    feat = layer.GetFeature(i)
    s = ""
    for field in fields:
        s +="{0}= {1} \t".format(field, feat.GetField(field))
    print(s)
    geo = feat.GetGeometryRef()
    print(geo)
```

```python
layer.ResetReading()
feDict = []
for feature in layer:
    f = {}
    for field in fields:
        f[field] = feature.GetField(field)
    geo = feature.GetGeometryRef()
    f["Shape"] = geo.ExportToWkt()
    feDict.append(f)
layer.ResetReading()
```

```python
import pandas
import matplotlib.pyplot as plt
```

```python
sonar = pandas.read_csv("./data/sonar.all-data.csv",header=None,prefix='X')
```

```python
sonar.rename(columns={'X60':'Label'}, inplace=True)
```

```python
sonar.Label = sonar.Label.astype('category')
```

```python
sonar.head()
```

# Andrews曲线
## 将每个样本的属性值转化为傅里叶序列的系数来创建曲线。
## 通过将每一类曲线标成不同颜色可以可视化聚类数据，属于相同类别的样本的曲线通常更加接近并构成了更大的结构。

# parallel_coordinates 平行坐标
## 平行坐标也是一种多维可视化技术。它可以看到数据中的类别以及从视觉上估计其他的统计量。
## 使用平行坐标时，每个点用线段联接。每个垂直的线代表一个属性。一组联接的线段表示一个数据点。可能是一类的数据点会更加接近。

```python
from pandas.plotting import andrews_curves, parallel_coordinates
fig, axes = plt.subplots(2,1,figsize=(12,8))
andrews_curves(sonar, 'Label', samples=50, linewidth=0.5, ax=axes[0],color=("r","g"))
axes[0].set_xticks([])
parallel_coordinates(sonar, 'Label', linewidth=0.5, ax=axes[1],color=("r","g"),
                     axvlines_kwds={'linewidth': 0.5, 'color': 'black', 'alpha':0.5})
axes[1].set_xticks([])
axes[1].margins(0.05)
pass
```

```python
from sklearn.manifold import MDS
mds = MDS(n_components=2)
mds_data = mds.fit_transform(sonar.iloc[:,:-1])
plt.scatter(mds_data[:, 0], mds_data[:, 1], c=sonar.Label.cat.codes, s=50);
```

```python
plt.figure(figsize=(9,6))
heatmap = plt.pcolor(sonar.corr(), cmap='jet')
plt.colorbar(heatmap)
#plt.gca().set_aspect('equal')
```

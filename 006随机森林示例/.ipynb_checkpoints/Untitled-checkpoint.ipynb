```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axisartist.axislines import SubplotZero
```

```python
x = np.linspace(0,1,1000)
```

```python
y = -2 * x*x + 2*x
```

```python
fig = plt.figure(1, (10, 6))
ax = SubplotZero(fig, 1, 1, 1)
fig.add_subplot(ax)
ax.axis["xzero"].set_visible(True)
ax.axis["xzero"].label.set_color('green')

ax.axis["x05"] = ax.new_floating_axis(nth_coord=0, value=0.5,axis_direction="bottom")
ax.axis["x05"].toggle(all=True)

ax.grid(True, linestyle='-.')
ax.axis["top",'right','left','bottom'].set_visible(False)
plt.plot(x,y,color="red")
plt.plot((0.5,0.5),(-0.1,0.55),color="green")
plt.text(0.51,0.56,"p = 0.5",fontdict={"fontsize":14})
pass
```

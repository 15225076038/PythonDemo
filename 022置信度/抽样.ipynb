```python
import numpy
import seaborn as sns 
```

## 创建10000个男生变量，72%设置为喜欢打篮球记为1，剩下的设置不喜欢打篮球，记为0

```python
play_bb = 0.72
boy = 10000
love = int(play_bb * boy)
dont_love = int(boy * (1 - play_bb))
loveb = numpy.ones(love)
nt_love = numpy.zeros(dont_love)
all_boy = numpy.hstack([loveb, nt_love])
```

## 进行10次抽样，随机找100个男生，统计有多少人喜欢打篮球

```python
mn = []
for i in range(10):
    sample = numpy.random.choice(all_boy, size=100)
    mn.append(numpy.mean(sample))
sns.distplot(mn,bins=10,
             kde_kws={"color":"r", "lw":3 }, 
             hist_kws={ "color": "b" }) 
print(mn)
```
